```python
tour_cap_nat_file_name = 'tour_cap_nat.tsv'
```

```python
path = os.path.join(os.getcwd(), tour_cap_nat_file_name)
tour_cap_df = pd.read_csv(path, sep='\t')
```

```python
col = tour_cap_df.columns[0]
```

```python
cols = col.split(',')
```

```python
tour_cap_df[cols] = tour_cap_df[col].str.split(",", expand=True)
```

```python
tour_cap_df = tour_cap_df.drop(col, axis=1)
```

```python
tour_cap_df
```

```python
tour_cap_df = tour_cap_df[(tour_cap_df['accommod'] == 'BEDPL') & (tour_cap_df['unit'] == 'NR') & (tour_cap_df['nace_r2'] == 'I551')]
```

```python
tour_cap_df
```

```python
cols.extend(['2016 '])
```

```python
tour_cap_df = tour_cap_df[cols]
```

```python
tour_cap_df = tour_cap_df.rename(columns={'geo\\time':'Country Code', '2016 ': 'Bed Number'})
```

## Task 2

```python
tin_file_name = 'tin00083.tsv'
```

```python
file_path = os.path.join(os.getcwd(), tin_file_name)
tin_df = pd.read_csv(file_path, sep='\t')
```

```python
col = tin_df.columns[0]
```

```python
cols = col.split(',')
```

```python
cols
```

```python
tin_df[cols] = tin_df[col].str.split(",", expand=True)
```

```python
tin_df = tin_df.drop(col, axis=1)
```

```python
tin_df
```

```python
tin_df = tin_df[tin_df['ind_type'] == 'IND_TOTAL']
```

```python
tin_df
```

```python
cols.extend(['2016 '])
```

```python
tin_df = tin_df[cols]
```

```python
tin_df
```

```python
drop_cols = (tin_df['geo\\time'].isin(['EA', 'EU27_2007', 'EU27_2020', 'EU28'])) | (tin_df['2016 '].str.endswith(': ')) | (tin_df['2016 '].str.endswith(' u')) | (tin_df['2016 '].str.endswith(' bu'))
```

```python
drop_cols
```

```python
tin_df = tin_df[~drop_cols]
```

```python
tin_df['2016 ']
```

```python
tin_df['2016 '] = tin_df['2016 '].str.rstrip(' b')
```

```python
tin_df = tin_df.rename(columns={'2016 ': 'Online Percentage', 'geo\\time':'Country Code'})
```

```python
df = tour_cap_df.merge(tin_df, left_on='Country Code', right_on='Country Code')
```

```python
df
```

```python
df = df[['Country Code', 'Bed Number', 'Online Percentage']]
```

```python
df
```

```python
df.to_csv('result.csv')
```

# Part 2

## Visualize

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = df.astype({'Bed Number':int, 'Online Percentage':int, 'Country Code':str})
```

```python
names = df['Country Code']
bed_values = df['Bed Number']
```

```python
plt.figure(figsize=(12,9), dpi=120)
plt.bar(names, bed_values)
plt.xlabel('Country Code')
plt.ylabel('Bed Number')
```

```python
names = df['Country Code']
percent_values = df['Online Percentage']
```

```python
plt.figure(figsize=(12,9), dpi=120)
plt.bar(names, percent_values)
plt.xlabel('Country Code')
plt.ylabel('Online Percentage')
```

```python
plt.figure(figsize=(12,9))
for i in df.index:
    plt.scatter(x=df.iloc[i]['Bed Number'], y=df.iloc[i]['Online Percentage'], label=df.iloc[i]['Country Code'])
plt.legend(loc='upper right')
# plt.title('Scatter of Country')
plt.xlabel('Bed Number')
plt.ylabel('Online Percentage')
```
